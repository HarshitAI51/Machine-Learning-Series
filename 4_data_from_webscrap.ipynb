### ⚠️ Note on Web Scraping in Google Colab

While performing web scraping in Google Colab, you may face issues even after using a User-Agent header. Some websites use protection services (such as Cloudflare or similar systems) that detect automated requests and block or modify the HTML response. As a result, expected tags or data may not appear in the parsed content.

This is not a coding error. It is a website-level restriction. In such cases, scraping may require advanced techniques (session handling, JavaScript rendering, or official APIs), or the data may not be accessible for scraping at all.


```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_India"

headers = {
    "User-Agent": "Mozilla/5.0"
}

html = requests.get(url, headers=headers).text
soup = BeautifulSoup(html, "lxml")
```

```python
soup.find('h1', id='firstHeading').text
```

```python
table = soup.find('table', class_='wikitable')
df = pd.read_html(str(table))[0]
df.head()
```

```python
for i in soup.find_all('h2'):
  print(i.text.strip())
```
