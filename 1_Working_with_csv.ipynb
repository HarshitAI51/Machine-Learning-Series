  1.Importing pandas

```python
import pandas as pd
```

2.Reading Locally

```python
df = pd.read_csv('data.csv')
df.head()
```

3.Reading using URL

```python
import requests
from io import StringIO
url = ""
headers = {"User-Agent": ""}
req = request.get(url,header = headers)
data = StringIO(req.text)
pd.read_csv(data)
```

4.Sep Parameter

```python
#when ever u use tsv file u have to change the sep to "," to \t because use
#different sep and names for just giving column names

tsv = pd.read_csv('data.csv',sep = '\t',names = ['sno', 'iq','age'])
tsv.head() #its show wired output because its csv
```

5.Index_col parameter

```python
pd.read_csv('data.csv',index_col = 'sno') #for change index column
```

6.header parameter

```python
pd.read_csv('data.csv',header = 1) #for change header some time headrer also come in index
```

7.use_cols parameter

```python
pd.read_csv('data.csv',usecols = ['sno','iq']) #only columns you need
```

8.Squeeze Parameter

```python
pd.read_csv('data.csv',usecols = ['sno'],squeeze = True) #only for use one colummn not use to much
```

9.skiprows parameter and nrows

```python
pd.read_csv('data.csv',skiprows = [0,1]) #for skip rows
pd.read_csv('data.csv',nrows = 10) #for read only 10 rows
```

10.Encoding Parameter

```python
pd.read_csv('data.csv',encoding = 'latin-1') #for encoding
```

11. Skip Bad Lines

```python
pd.read_csv('data.csv',error_bad_lines = False) #for skip bad lines or in which the lines are not correct
```

12.dtypes Parameter

```python
pd.read_csv('data.csv',dtype = {'iq':float}) #for change the data type
```

13.Handling Dates

```python
pd.read_csv('data.csv',parse_dates = ['date']) #for change string in date format
```

14.Convertors

```python
def rename(name):
    if name == "group1":
        return "g1"
    else:
        return name
pd.read_csv('data.csv',converters = {'group':rename}) #for change the values
```

15.na_values parameter

```python
pd.read_csv('data.csv',na_values = ['male']) #for change parametr to Nan or empty
```

16.Loading Huge datasheet in chunks

```python
dfs = pd.read_csv('data.csv',chunksize = 1000) #for load data in chunks or means simply make parts of data

for chunks in dfs:
    print(chunks.shape)
```
